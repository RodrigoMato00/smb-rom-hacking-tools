```python
from nes import NES

import matplotlib.pyplot as plt
%matplotlib inline

mynes = NES("./roms/Super Mario Bros. (Japan, USA).nes", 
            verbose=False, headless=True)
buffer = mynes.run_frame_headless(run_frames=60)
plt.figure(figsize=(7.5, 7))
plt.imshow(buffer)
```

```python
buffer = mynes.run_frame_headless(run_frames=180, 
                                  controller1_state=[0,0,0,1,0,0,0,0])
plt.figure(figsize=(7.5, 7))
plt.imshow(buffer)
```

```python
buffer = mynes.run_frame_headless(run_frames=30, 
                                  controller1_state=[1,0,0,0,0,0,0,0])
plt.figure(figsize=(7.5, 7))
plt.imshow(buffer)
```
